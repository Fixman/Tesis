```python
%pylab inline

from __future__ import print_function
from __future__ import division

import numpy
import matplotlib
import seaborn
import scipy

from matplotlib import pyplot
from scipy import stats
```

```python
x = numpy.linspace(0, 1, 50)

params = {
    '0 poor contacts, 3 rich contacts': (1, 4),
    '4 poor contact, 1 rich contacts': (5, 2),
    '6 poor contacts, 6 rich contacts': (7, 7)
}

Y = {
    name: map(lambda x: stats.beta.pdf(x, a, b), x)
    for name, (b, a) in params.iteritems()
}
```

```python
%pwd
```

```python
pyplot.rcParams['figure.figsize'] = 15.33, 6.67

for name, (b, a) in params.iteritems():
    y = map(lambda x: stats.beta.pdf(x, a, b), x)
    line = pyplot.plot(x, y, label = name)
    
    q = stats.beta.ppf(.05, a, b)
    pyplot.gca().add_patch(
        matplotlib.patches.Polygon(
            [(0, 0)] + [(f, g) for f, g in zip(x, y) if f <= q] + [(q, 0)],
            edgecolor = 'black',
            alpha = .75,
            facecolor = line[0].get_color()
        )
    )
    
pyplot.legend(loc = 0)

# pyplot.title('Value of the 5th percentile for different neighbours\' socioeconomic levels')
pyplot.ylabel('Probability Density Function')

pyplot.savefig('../Asonam2016/figures/Beta_example.png', dpi = 500, bbox_inches='tight')
```
