```python
%pylab inline

from __future__ import print_function
from __future__ import division

import numpy
import matplotlib
import seaborn
import scipy

from matplotlib import pyplot
from scipy import stats
```

```python
x = numpy.linspace(0, 1, 200)

params = [
    (1, 4),
    (5, 2),
    (21, 6),
    (7, 7),
]

Y = {
    '{} rich contacts, {} poor contacts'.format(a - 1, b - 1): map(lambda x: stats.beta.pdf(x, a, b), x)
    for b, a in params
}
```

```python
pyplot.rcParams['figure.figsize'] = 10, 4

for b, a in params:
    y = map(lambda x: stats.beta.pdf(x, a, b), x)
    line = pyplot.plot(x, y, label = '{} rich contacts, {} poor contacts'.format(a - 1, b - 1))
    
    q = stats.beta.ppf(.05, a, b)
    pyplot.gca().add_patch(
        matplotlib.patches.Polygon(
            [(0, 0)] + [(f, g) for f, g in zip(x, y) if f <= q] + [(q, 0)],
            edgecolor = 'black',
            alpha = .75,
            facecolor = line[0].get_color()
        )
    )
    
pyplot.legend(loc = 0)

# pyplot.title('Value of the 5th percentile for different neighbours\' socioeconomic levels')
pyplot.ylabel('Probability Density Function')

pyplot.savefig('../figures/Beta_example.png', dpi = 500, bbox_inches='tight')
```
